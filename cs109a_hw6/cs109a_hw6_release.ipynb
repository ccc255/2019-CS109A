```python
### cs109Test(test_1.1a) ###
np.random.seed(42)
depths = list(range(1, 21))

def calc_meanstd(X_train, y_train, depths):
    #Find optimal depth of trees
    cvmeans = {}
    cvstds = {}
    train_scores = {}
    for i in depths:
        model = DecisionTreeClassifier(max_depth=i)
        score = cross_val_score(estimator=model, X=X_train, y=y_train, cv=5, n_jobs=-1)
        cvmeans[i] = score.mean()
        cvstds[i] = score.std()
        train_scores[i] = score
    return cvmeans, cvstds, train_scores
```

**1.3** What is the mechanism by which limiting the depth of the tree avoids over-fitting? What is one downside of limiting the tree depth?

Limiting the depth of the tree avoids overfitting by limiting the number of splits that we have to make with small subsamples of data. We can think of a deep tree as overfitting in a similar way to how a K-Nearest Neighbors model with low k would overfit (since the prediction or splitting criteria must be determined from only very select group of data points). In the lower down nodes of the tree, there is not much data used to make the split at each leaf and so the splits may be influenced by the noise too much and not be generalizable to the test set (this means that the lower down nodes of the tree may be contributing to overfitting).

One downside of limiting the tree depth is the model may not be complex enough and thus have high bias and low variance.

```python
### cs109test(test_2.2) ### 
np.random.seed(42)
def bagger(n_trees: int, tree_depth: int,  random_seed=0) -> (pd.DataFrame, pd.DataFrame, list):
    """A function that takes as 
    
    Inputs:
      n_tres
      tree_depth 
      a random_seed (default =0)
    
    Returns:
      bagging_train dataframe (as described above)
      bagging_test dataframe (as described above)
      bagging_models every trained model for each bootstrap (you will need this in Q3.2)
    """
    #Creating model
    np.random.seed(random_seed)
    model = DecisionTreeClassifier(max_depth=tree_depth)

    #Initializing variables
    predictions_train = np.zeros((data_train.shape[0], n_trees))
    predictions_test = np.zeros((data_test.shape[0], n_trees))
    
    bagging_models = []
    #Conduct bootstraping iterations
    for i in range(n_trees):
        temp = data_train.sample(frac=1, replace=True)
        response_variable = temp['class']
        temp = temp.drop(['class'], axis=1)

        bagging_models.append(model.fit(temp, response_variable))  
        predictions_train[:,i] = model.predict(X_train)   
        predictions_test[:,i] = model.predict(X_test)

    #Make Predictions Dataframe
    columns = ["bootstrap model "+str(i+1) +"'s prediction" for i in range(n_trees)]
    bagging_train_df = pd.DataFrame(predictions_train, columns=columns)
    bagging_test_df = pd.DataFrame(predictions_test, columns=columns)
    return bagging_train_df, bagging_test_df, bagging_models
```

**2.4** We want to know how the number of bootstraps affects our bagging ensemble's performance. Use the `running_predictions` function (given below) to get the model's accuracy score when using only 1,2,3,4,... of the bootstrapped models. Make a plot of training and test set accuracies as a function of number of bootstraps.
** use the depth that you used above **

On your plot, also include horizontal lines for two baselines:
- the test accuracy of the best model from question 1
- the test accuracy of a single tree with the tree depth you chose in 2.1, trained on the full training set.


```python
np.random.seed(42)
def running_predictions(prediction_dataset: np.ndarray, targets: np.ndarray) -> np.ndarray:
    """A function to predict examples' class via the majority among trees (ties are predicted as 0)
    
    Inputs:
      prediction_dataset - a (n_examples by n_sub_models) dataset (not a dataframe), where each entry [i,j] is sub-model j's prediction
          for example i
      targets - the true class labels
    
    Returns:
      a vector where vec[i] is the model's accuracy when using just the first i+1 sub-models
    """
    
    n_trees = prediction_dataset.shape[1]
    
    # find the running percentage of models voting 1 as more models are considered
    running_percent_1s = np.cumsum(prediction_dataset, axis=1)/np.arange(1,n_trees+1)
    
    # predict 1 when the running average is above 0.5
    running_conclusions = running_percent_1s > 0.5
    
    # check whether the running predictions match the targets
    running_correctnesss = running_conclusions == targets.reshape(-1,1)
    
    return np.mean(running_correctnesss, axis=0)
    # returns a 1-d series of the accuracy of using the first n trees to predict the targets
```

```python
# plotting code
np.random.seed(42)
y_train_running_predictions = running_predictions(np.array(bagging_train_df), np.array(y_train))
y_test_running_predictions = running_predictions(np.array(bagging_test_df), np.array(y_test))
x_running_predictions = list(range(1,56))

plt.ylabel("Model accuracy using the first n trees to predict")
plt.xlabel("The number n trees used to make predictions")
plt.title('Bagging performance as a function of number of bootstrap samples')
plt.plot(x_running_predictions, y_train_running_predictions, 'b-', marker='o',label = 'Running accuracy in training set')
plt.plot(x_running_predictions, y_test_running_predictions, 'b-', marker='x', label = 'Running accuracy in testing set')
plt.axhline(best_cv_tree_test_score, color = 'grey', label = 'Tree depth of 5 (best tree depth)')
plt.axhline(overfit_cv_tree_test_score, color = 'r', label = 'Tree depth of 20 (overfitting)')
plt.legend()
plt.show()
```

**5.1** How do boosting and bagging relate: what is common to both, and what is unique to each?



Both bagging and boosting are ensemble methods of learning decision trees, meaning that both share the idea of combining multiple trees to learn a better model (ideally improving the accuracy). However, bagging starts with deep trees and boosting starts with simple trees. Boosting is an iterative process (where the next model directly depends on the preceeding one) while bagging is not iterative in the sense that trees are independent of each other. Boosting weights misclassified observations more heavily than correctly classified ones in order to learn the model while bagging does not treat misclassified versus correctly classified observations differently in the learning process. Because bagging starts with complex trees, it reduces variance. Boosting, because it starts with simple trees, reduces bias.

**5.4** Which technique, boosting or bagging, is better suited to parallelization, where you could have multiple computers working on a problem at the same time?


Bagging is better suited to parallelization as each model can be constructed using parallelized bootstrapping, but boosting can't be done in parallel because each successive tree is dependent on the classification mistakes made in the previous tree (boosting is an interative process). So, for any specific depth, boosting is not suited for parallelization.
